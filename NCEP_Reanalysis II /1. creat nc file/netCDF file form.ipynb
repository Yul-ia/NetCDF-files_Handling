```python
from netCDF4 import Dataset
# import netCDF4 as nc
import os
os.chdir('')
import xarray as xr
```

## CMAP

```python
url = 'http://psl.noaa.gov/thredds/dodsC/Datasets/cmap/enh/precip.mon.mean.nc'
# ds = nc.Dataset(url)
# ds.to_netcdf('precip.mon.mean.nc')
ds = xr.open_dataset(url)

lon = ds.lon
lat = ds.lat
time = ds.time
precip = ds.precip
#%%
## cmap
ds_new = xr.Dataset(
    data_vars= dict(
        precip = (['time','lat','lon'],precip.data),
    ),
    coords = dict(
        time = time.data,
        lon = lon.data,
        lat = lat.data
    )
)

# 저장
ds_new.to_netcdf('precip.mon.mean.nc') #cmap저장
```

## NCEP

```python
## ncep_ air_temperature
url_nep_airtemp = 'http://psl.noaa.gov/thredds/dodsC/Datasets/ncep.reanalysis2/Monthlies/gaussian_grid/air.2m.mon.mean.nc'
ds_airtemp = xr.open_dataset(url_nep_airtemp)

lon = ds_airtemp.lon
lat = ds_airtemp.lat
time = ds_airtemp.time
level= ds_airtemp.level
air = ds_airtemp.air

ds_new_airtemp = xr.Dataset(
    data_vars= dict(
        air = (['time','level','lat','lon'],air.data),
    ),
    coords= dict(
        level = level.data,
        time = time.data,
        lon = lon.data,
        lat = lat.data
    )
)

ds_new_airtemp.to_netcdf('air.2m.mon.mean.nc') # 저장
```

```python
## ncep _precipitation
url_nep_prate = 'http://psl.noaa.gov/thredds/dodsC/Datasets/ncep.reanalysis2/Monthlies/gaussian_grid/prate.sfc.mon.mean.nc'
ds_prate = xr.open_dataset(url_nep_prate)

# ds_prate

lon = ds_prate.lon
lat = ds_prate.lat
time = ds_prate.time
prate = ds_prate.prate
 
ds_new_prate = xr.Dataset(
    data_vars= dict(
        prate = (['time','lat','lon'], prate.data),
    ),
    coords= dict(
        time = time.data,
        lon = lon.data,
        lat = lat.data
    )
)

ds_new_prate.to_netcdf('prate.sfc.mon.mean.nc') #저장
```
